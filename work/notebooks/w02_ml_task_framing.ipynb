## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

```python
## 1. My lane as an ML task (type)

My project is a **ranking and scoring task** with a classification component.

The main goal is to create a ranked list of content pages that should be reviewed first for refresh opportunities. The model assigns a score or probability to each page based on observable signals such as impressions, clicks, CTR, sessions, average position, engagement rate, and content age.

It also uses classification because the model predicts whether a content page shows a decline pattern (`is_declining_label`). However, the final business output is not only a yes/no prediction — it is a prioritized queue of pages ranked by review importance.

Therefore, the primary ML task is **ranking/scoring**, supported by a classification model.
```

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

```python
## 2. Target or proxy

The target used in this project is `is_declining_label`.

This label represents whether a content page shows a declining trend based on the observed field `trend_direction`.

The label is created using a defined rule:

`is_declining_label = trend_direction == "down"`

This means the target is a **proxy label**, not a direct future outcome. It is a simplified representation of content decline based on the current observed data window.

The label does not prove that a page will continue declining in the future. A stronger future version of this project could define a prediction task such as using the previous 90 days of signals to predict decline or recovery in the next 30 days.

For this internship project, the proxy label helps build and evaluate a ranking system while keeping the limitations clear.
import pandas as pd

# Load processed feature data
df = pd.read_csv("../../data/processed/refresh_feature_vector.csv")

# Check target column
print("Target column exists:", "is_declining_label" in df.columns)

# Show target distribution
print("\nTarget distribution:")
print(df["is_declining_label"].value_counts())

# Percentage of declining pages
print("\nDeclining rate:")
print(round(df["is_declining_label"].mean() * 100, 2), "%")
```

## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
## 3. Success metric

The main success metric for this project is **Precision@50**.

Precision@50 measures how many of the top 50 pages recommended by the model are actually relevant according to the defined decline label.

This metric matches the real business decision because a content team usually has limited time and can only review a fixed number of pages first.

A higher Precision@50 means that the ranked queue contains more useful recommendations.

The goal is to achieve a Precision@50 higher than the baseline scoring method, showing that the ML model provides better prioritization than simple hand-written rules.

A good result is not only a high score but also a ranking that provides explainable reason codes for why pages were selected.

import json

# Load model results
with open("../../outputs/model_results.json", "r") as f:
    results = json.load(f)

# Display available metrics
print(results.keys())

print("\nModel evaluation results:")
print(results)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

```python
## 5. Why ML beats a fixed rule here

A fixed rule-based system can identify simple patterns, but content performance depends on many signals interacting together.

For example, a page may have:
- high impressions but low CTR,
- good search position but declining engagement,
- old content but still strong performance,
- recent content with weak visibility.

These combinations are difficult to capture using only manually written if-statements.

Machine learning can learn patterns across multiple features at the same time and identify combinations of signals that are associated with declining content.

In this project, the baseline provides a transparent rule-based score, while ML models such as Random Forest learn more complex relationships between search visibility, engagement, freshness, and content signals.

The goal is not to replace human judgment, but to improve the prioritization of pages for review.

import json

# Load model results
with open("../../outputs/model_results.json", "r") as f:
    results = json.load(f)

# Print evaluation metrics
print(results)
```
