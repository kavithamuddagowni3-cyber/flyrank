## 1. Question

### Research question

Which content pages should be reviewed first based on observable search and engagement signals?

### Decision supported

This project supports content teams in deciding where to spend limited review time by creating a ranked list of pages that may need attention.

The system helps answer:

"Which pages should a reviewer investigate first, and why?"

The output is a decision-support ranking, not an automatic content action. Human reviewers still decide whether a page should be refreshed, improved, monitored, or left unchanged.

## 3. Methodology

*Assumptions, features, label definition, baseline, validation design, leakage checks.*

```python
## 3. Methodology

### Assumptions

This project assumes that observable content and search performance signals can help identify pages that deserve earlier review.

The model is designed for prioritization, not for guaranteeing that a specific action will improve performance.

---

### Features

The model uses observable signals available before prediction, including:

- impressions and clicks;
- CTR;
- average search position;
- sessions;
- engagement signals;
- content age and freshness;
- content structure measurements;
- categorical content attributes.

These features represent measurable patterns in content performance.

---

### Label definition

The prediction target is:

`is_declining_label`

The label is defined using the available trend signal:

- `trend_direction == "down"` → declining label (1)
- otherwise → non-declining label (0)

This is a proxy label based on an observed rule, not a direct measure of whether a content refresh succeeds.

---

### Baseline approach

A transparent rule-based baseline was created before applying machine learning.

The baseline combines signals such as:

- visibility;
- freshness risk;
- position opportunity;
- content depth.

The purpose of the baseline is to provide a simple benchmark that the ML model must improve upon.

---

### Model approach

A Random Forest classifier was selected because:

- it can capture non-linear relationships between multiple content signals;
- it handles mixed feature patterns effectively;
- it provides feature importance for interpretation.

The model output is converted into a ranked review queue based on risk/opportunity scores.

---

### Validation design

The model was evaluated using a client-holdout split.

This means:

- complete clients are separated between training and testing;
- the model is tested on content from clients it has not seen before.

This provides a more realistic evaluation than randomly splitting individual pages.

Primary evaluation metric:

**Precision@50**

This matches the real decision because content teams usually review a limited number of pages first.

---

### Leakage checks

A leakage audit was performed to ensure the model only uses information available before prediction.

Checks included:

- removing product-generated fields such as `health_score`, `priority_score`, and `action_type`;
- excluding future performance metrics;
- ensuring the target label is not used as an input feature;
- removing identifiers such as client/content IDs from predictive features.

The final model uses only observable signals and does not rely on hidden decisions or future information.
```

## 4. Results (vs baseline)

*Model vs baseline on the same split. The honest table.*

```python
## 4. Results (vs baseline)

The machine learning model was compared against the transparent baseline using the same evaluation setup and the same ranking metric.

### Evaluation metric

**Precision@50**

Precision@50 measures how many of the top 50 recommended pages matched the defined decline label.

This metric was selected because the practical use case is prioritizing a limited number of pages for human review.

### Model comparison

| Method | Approach | Precision@50 |
|---|---|---:|
| Baseline rules | Hand-written scoring rules based on visibility, freshness, position, and content signals | 0.240 |
| Random Forest | Machine learning ranking model using observable content and search signals | 0.740 |

### Interpretation

The Random Forest model achieved higher Precision@50 than the fixed baseline on the evaluated split.

This suggests that combining multiple observable signals through a learned model can improve the prioritization of pages for review compared with a simple manually defined rule.

The result should be interpreted as improved ranking performance, not as proof that a recommended page will recover after a content change.
```

## 5. Limitations

*What this work cannot claim.*

```python
## 5. Limitations

This project is a decision-support system for prioritizing content review. It has several important limitations.

### No causal proof

The model identifies pages that show patterns associated with decline risk or review opportunity. It cannot prove that refreshing or changing a page will cause traffic or engagement improvement.

### Not predicting search engine rankings

The model does not predict Google rankings or search engine algorithms. It only learns relationships between the provided observable content and performance signals.

### Proxy label limitation

The target label is based on an observed trend rule (`trend_direction == "down"`), which acts as a proxy for decline. It is not a direct measurement of successful or unsuccessful content updates.

### Human review is required

The ranked queue helps reviewers decide where to investigate first. A human must consider additional context such as business goals, seasonality, consolidation, and content quality before taking action.

### Data limitations

The dataset is anonymized and contains an unbalanced history across content items and clients. Some measurements may be missing because tracking systems were not available for every period.

### Generalization limits

The results are based on the provided internship dataset and validation setup. Performance may change when applied to new datasets, clients, or different time periods.
```

## 6. Ranked recommendations

*The action playbook output — the paper's recommendations section.*

```python
## 6. Ranked Recommendations

The final output of this project is a ranked action queue that helps content teams decide which pages to review first.

The ranking combines model scores with observable signals and provides reason codes so that reviewers can understand why a page was prioritized.

### Recommendation approach

Pages are prioritized based on:

- predicted decline risk from the machine learning model;
- search visibility and demand signals;
- engagement performance;
- content freshness;
- opportunity signals.

### Example action categories

| Recommendation | When it applies |
|---|---|
| Refresh content | Pages with visibility, decline signals, and enough demand to justify review |
| Improve search appearance | Pages with strong impressions but lower-than-expected click performance |
| Investigate engagement | Pages receiving traffic but showing weaker engagement signals |
| Monitor performance | Pages with signals that need more evidence before action |
| Protect successful pages | Pages performing well and showing strong visibility |

### Reason codes

Each recommendation includes a reason code to make the ranking explainable.

Examples:

- `model_decline_risk` — model identifies a higher probability of decline pattern;
- `visible_model_opportunity` — page has meaningful visibility and improvement opportunity;
- `ctr_review_candidate` — page receives impressions but has lower click-through performance;
- `engagement_review_candidate` — page has traffic but weaker engagement indicators.

### Human decision process

The ranked list is not an automatic action system.

A content reviewer should verify:

- whether the decline is real or caused by seasonality;
- whether another page has absorbed traffic;
- whether the recommended change matches user intent;
- whether updating the page is worth the effort.

The goal is to help teams spend review time on the pages with the strongest evidence first.
```
